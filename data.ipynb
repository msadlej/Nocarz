```python
import pandas as pd
```

```python
users = pd.read_csv("users/users.csv")
users
```

```python
dropped_users = users.dropna(subset=['id'])
dropped_users
```

```python
# Calculate amount and percentage of rows dropped
total_rows = len(users)
dropped_rows = total_rows - len(dropped_users)
drop_percentage = (dropped_rows / total_rows) * 100 if total_rows > 0 else 0

print(f"Total rows in original dataset: {total_rows}")
print(f"Rows dropped due to missing 'id': {dropped_rows}")
print(f"Percentage of rows dropped: {drop_percentage:.2f}%")
```

```python
reviews = pd.read_csv("reviews/reviews.csv")
reviews
```

```python
dropped_reviews = reviews.dropna(subset=['listing_id', 'comments'])
dropped_reviews
```

```python
# Calculate amount and percentage of rows dropped
total_rows = len(reviews)
dropped_rows = total_rows - len(dropped_reviews)
drop_percentage = (dropped_rows / total_rows) * 100 if total_rows > 0 else 0

print(f"Total rows in original dataset: {total_rows}")
print(f"Rows dropped due to missing 'listing_id' or 'comments': {dropped_rows}")
print(f"Percentage of rows dropped: {drop_percentage:.2f}%")
```
